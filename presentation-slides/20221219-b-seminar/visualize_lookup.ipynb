```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
probs = [0.2,0,0.1,0,0,0.3,0,0.4]
```

```python
df = pd.DataFrame({
    "P(X=k)":probs
})
df["k"] = [k for k in range(len(df))]
df = df.set_index("k",drop=True)  # df= df.reindex(columns=["k","P(X=k)"])
df["P(X<=k)"] = df["P(X=k)"].cumsum()
df
```

```python
fig,axs = plt.subplots(1,2,figsize=(15,8),sharey = "all")
left = df["k"]
height = df["P(X=k)"]
label = df["k"]
axs[0].bar(left, height,tick_label=label, align="center")
axs[0].set_title('PMF')
axs[0].set_xlabel('k value')
axs[0].set_ylabel('P(X=K) value')
axs[0].grid()

left = df["k"]
height = df["P(X<=k)"]
label = df["k"]
axs[1].bar(left, height,tick_label=label, align="center")
axs[1].set_title('CMF')
axs[1].set_xlabel('k value')
axs[1].set_ylabel('P(X<=K) value')
axs[1].grid()
plt.show()
```
